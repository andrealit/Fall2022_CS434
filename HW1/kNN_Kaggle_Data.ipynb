### Create all Functions We Need

```python
# Euclidian distance
def euclidian_distance(row1, row2):
    distance = 0.0
    for i in range(len(row1) - 1):
        distance += (row1[i] - row2[i])**2
    return np.sqrt(distance)
```

```python
def get_nearest_neighbors(example_set, query, k):
    distances = list()
    idx = 0
    for example_row in example_set:
        dist = euclidian_distance(query, example_row)
        distances.append((idx, example_row, dist))
        idx += 1;
    
    distances.sort(key=lambda tup: tup[2])
    neighbors_idx = list()
    for i in range(k):
        neighbors_idx.append(distances[i][0])
    return neighbors_idx
```

```python
def knn_classify_point(examples_X, examples_y, query, k):
    # get the kNN list base on k
    #print("knn_classify_point(examples_X) shape", examples_X.shape)
    #print("knn_classify_point(examples_y) shape", examples_y.shape)
    #print("knn_classify_point(query) shape", query.shape)
    neighbors = get_nearest_neighbors(examples_X, query, k)
    # print(neighbors)
    # print(examples_y)
    
    # put the output train example_y in this
    output_nearest_neighbors = []
    # output_nearest_neighbors = []
       
    # get the output of nearest neighbors based on k value
    # NEED FURTHER DEBUGGING
    for neighbor_idx in neighbors:
        output_nearest_neighbors.append(examples_y[neighbor_idx][0])
    
    # output_nearest_neighbors = np.array(output_nearest_neighbors)
    #print(output_nearest_neighbors)
    # print(type(output_nearest_neighbors))
    
    
    # print("output_nearest_neighbors = ",output_nearest_neighbors)
    # get the maximum vote from the kNN output_nearest_neighbors
    predicted_label = max(set(output_nearest_neighbors), key=output_nearest_neighbors.count)
    
   
    return predicted_label
```

```python
def predict(examples_X, examples_y, queries_X, k): 
    # For each query, run a knn classifier
    predicted_y = [knn_classify_point(examples_X, examples_y, query, k) for query in queries_X]
    return predicted_y
```

```python
def cross_validation(train_X, train_y, num_folds, k):
    # Set some variables to use
    avg_val_acc = 0
    varr_val_acc = 0
    folds_container_X = []
    folds_container_y = []
    accuracy = []
    
    # Break data set into number of folds
    # and work on each fold
    for i in range(num_folds):
        folds_container_X.append(np.array_split(train_X, num_folds)[i])
        print("X", len(folds_container_X[i]))
        folds_container_y.append(np.array_split(train_y, num_folds)[i])
        print("y", len(folds_container_y[i]))
        queries = folds_container_X[i]
        
    # Get each data set into the fold_containers for train_X and train_y
    for i in range(num_folds):
        folds_container_X.append(np.array_split(train_X, num_folds)[i])
        print("X", len(folds_container_X[i]))
        folds_container_y.append(np.array_split(train_y, num_folds)[i])
        print("y", len(folds_container_y[i]))

    # looping Idea for i
    #     the train_X data has to coming from folds_container_X[not i]
    #     the train_y data has to coming from folds_container_y[not i]
    # if i == 0 then collect train data and label that greater than i
    # if i == num_folds-1 then collect train data and label from i = 0 to i-1 
    # if (i != 0) and (i < num_folds-1) then collect train data and label from i = 0 to i-1 and
    #                                      collect train data and label from i+1 to num_folds-2

    for i in range(num_folds):
        # For i == 0 then collect train data and label that greater than i
        if (i == 0):
            print("====================================", i)
            print("i == 0")
            queries = folds_container_X[i]
            print("queries of index ", i, " data rows = ",len(queries))
            # Stack data for train_fold_X into list
            print("Stack data for train_fold_X into list for ", i)
            train_fold_X = []
            train_fold_y = []
            p = i + 1
            for p in range(1,num_folds):
                for x in folds_container_X[p]:
                    train_fold_X.append(x)
                for y in folds_container_y[p]:
                    train_fold_y.append(y)
            print("Size of train_fold_X ", len(train_fold_X))
            # Convert train_fold_X list into numpy array
            train_fold_X = np.array(train_fold_X)
            print("Size of train_fold_y ", len(train_fold_y))
            # Convert train_fold_y list into numpy array
            train_fold_y = np.array(train_fold_y)
            # Call Prediction Function
            #k = 1
            pred3 = predict(train_fold_X, train_fold_y, queries, k)
            print("pred3 of index ", i, " data rows = ",len(pred3))
            # Compare Prediction to Actual 
            count = 0
            for v in range(len(pred3)):
                if (pred3[v] == folds_container_y[i][v]):
                    count +=1
            print("Correct Prediction = ", count)
            # Calculate Accuracy
            total = len(queries)
            print("Total = ", total)
            print("Count = ", count)
            print("Accuracy is Correct Prediction / Total = ", (count/total))
            accuracy.append((count/total))
            print(accuracy) 
    
        # for (i != 0) and (i < num_folds-1) then collect train data and label from i = 0 to i-1 and
        #                                        collect train data and label from i+1 to num_folds-2
        elif ((i != 0) and (i < num_folds-1)):
            print("====================================", i)
            print("(i != 0) and (i < num_folds-1)")
            queries = folds_container_X[i]
            print("queries of index ", i, " data rows = ",len(queries))
            # Stack data for train_fold_X into list
            print("Stack data for train_fold_X into list for ", i)
            train_fold_X = []
            train_fold_y = []
            p = 0
            for p in range(0,(num_folds-2)):
                for x in folds_container_X[p]:
                    train_fold_X.append(x)
                for y in folds_container_y[p]:
                    train_fold_y.append(y)
            print("Size of train_fold_X ", len(train_fold_X))
            # Convert train_fold_X list into numpy array
            train_fold_X = np.array(train_fold_X)
            print("Size of train_fold_y ", len(train_fold_y))
            # Convert train_fold_y list into numpy array
            train_fold_y = np.array(train_fold_y)
            # Call Prediction Function
            #k = 1
            pred3 = predict(train_fold_X, train_fold_y, queries, k)
            # Compare Prediction to Actual 
            count = 0
            for v in range(len(pred3)):
                if (pred3[v] == folds_container_y[i][v]):
                    count +=1
            print("Correct Prediction = ", count)
            # Calculate Accuracy
            total = len(queries)
            print("Total = ", total)
            print("Count = ", count)
            print("Accuracy is Correct Prediction / Total = ", (count/total))
            accuracy.append((count/total))
            print(accuracy)
        # For i == num_folds-1 then collect train data and label from i = 0 to i-1 
        elif (i == num_folds-1):
            print("====================================", i)
            print("i == num_folds-1 of ", i)
            queries = folds_container_X[i]
            print("queries of index ", i, " data rows = ",len(queries))
            # Stack data for train_fold_X into list
            print("Stack data for train_fold_X into list for ", i)
            train_fold_X = []
            train_fold_y = []
            p = 0
            for p in range(0,(num_folds-1)):
                for x in folds_container_X[p]:
                    train_fold_X.append(x)
                for y in folds_container_y[p]:
                    train_fold_y.append(y)
            print("Size of train_fold_X ", len(train_fold_X))
            # Convert train_fold_X list into numpy array
            train_fold_X = np.array(train_fold_X)
            print("Size of train_fold_y ", len(train_fold_y))
            # Convert train_fold_y list into numpy array
            train_fold_y = np.array(train_fold_y)
            # Call Prediction Function
            #k = 1
            pred3 = predict(train_fold_X, train_fold_y, queries, k)
            # Compare Prediction to Actual 
            count = 0
            for v in range(len(pred3)):
                if (pred3[v] == folds_container_y[i][v]):
                    count +=1
            print("Correct Prediction = ", count)
            # Calculate Accuracy
            total = len(queries)
            print("Total = ", total)
            print("Count = ", count)
            print("Accuracy is Correct Prediction / Total = ", (count/total))
            accuracy.append((count/total))
            print(accuracy)
    
    avg_val_acc = np.average(accuracy)
    varr_val_acc = np.var(accuracy)
    
    return avg_val_acc, varr_val_acc
```
